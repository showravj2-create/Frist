```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple friendship network
G = nx.Graph()
G.add_edges_from([
    ("Alice", "Bob"),
    ("Alice", "Charlie"),
    ("Bob", "David"),
    ("Charlie", "David"),
    ("David", "Eve"),
    ("Eve", "Frank")
    ])

# Draw the network
nx.draw(G, with_labels=True, node_color="lightblue", node_size=800)
plt.show()

# Find most connected node
degrees = dict(G.degree())
most_connected = max(degrees, key=degrees.get)
print(f"Most connected person: {most_connected}")
```
